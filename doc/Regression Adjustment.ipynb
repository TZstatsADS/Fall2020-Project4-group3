```python
### load required packages
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegressionCV
```

# 1. Low-dimensional data

```python
# Reading in data

lowDim_data = pd.read_csv('../data/lowDim_dataset.csv')
lowDim_data
```

### 1.1 Estimating propensity score

```python
X = lowDim_data.iloc[:, 2:]
A = lowDim_data['A']
Y = lowDim_data['Y']
clf = LogisticRegression(random_state = 0, solver = 'lbfgs', penalty = 'l2', C = 1).fit(X, A)
PS = clf.predict_proba(X)[:,1]

len(PS)
```

### 1.2 Regression adjustment

```python
# Merging A and PS
a = []
for i in range(0, len(A)):
    c = []
    c.append(A[i])
    c.append(PS[i])
    a.append(c)

# Linear Regression
model = LinearRegression().fit(a, Y)
```

```python
# ATE is the coefficient associated with indicator variable A

ATE = model.coef_[0]
ATE
```

# 2. High-dimensional data

```python
# Reading in data

highDim_data = pd.read_csv('../data/highDim_dataset.csv')
highDim_data
```

### 2.1 Estimating propensity score

```python
X = highDim_data.iloc[:, 2:]
A = highDim_data['A']
Y = highDim_data['Y']
clf = LogisticRegression(random_state = 0, solver = 'lbfgs', max_iter=10000, penalty = 'l2', C = 1).fit(X, A)
PS = clf.predict_proba(X)[:,1]
```

### 2.2 Regression adjustment

```python
# Merging A and PS
a = []
for i in range(0, len(A)):
    c = []
    c.append(A[i])
    c.append(PS[i])
    a.append(c)

# Linear Regression
model = LinearRegression().fit(a, Y)
```

```python
# ATE is the coefficient associated with indicator variable A

ATE = model.coef_[0]
ATE
```
